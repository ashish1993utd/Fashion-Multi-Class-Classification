# STEP #1: PROBLEM STATEMENT

Fashion training set consists of 70,000 images divided into 60,000 training and 10,000 testing samples. Dataset sample consists of 28x28 grayscale image, associated with a label from 10 classes. 

The 10 classes are as follows:  
0 => T-shirt/top
1 => Trouser
2 => Pullover
3 => Dress
4 => Coat
5 => Sandal
6 => Shirt
7 => Sneaker
8 => Bag
9 => Ankle boot

Each image is 28 pixels in height and 28 pixels in width, for a total of 784 pixels in total. Each pixel has a single pixel-value associated with it, indicating the lightness or darkness of that pixel, with higher numbers meaning darker. This pixel-value is an integer between 0 and 255.

# STEP #4: TRAINING THE MODEL

```python
# Prepare the training and testing dataset 
# Divding by 255 to scale the image features between 0 and 1 : Classic Normalization
X_train = training[:,1:]/255
y_train = training[:,0]

X_test = testing[:,1:]/255
y_test = testing[:,0]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_validate, y_train, y_validate = train_test_split(X_train, y_train, test_size = 0.2, random_state = 2)
```

```python
# * unpack the tuple
# X_train is in the form of an array, we need it to be in 28 by 28 by 1 matrix form
X_train = X_train.reshape(X_train.shape[0], *(28, 28, 1))
# Forms a matrix with rows as number of rows in X_train consisting a 28 * 28 matrix

X_test = X_test.reshape(X_test.shape[0], *(28, 28, 1))
X_validate = X_validate.reshape(X_validate.shape[0], *(28, 28, 1))
```

```python
# open source Neural network library
import keras
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
from keras.optimizers import Adam
from keras.callbacks import TensorBoard
```

```python
cnn_model = Sequential()

# Adding different layers to the sequential model
# Try 32 filters first then 64

# Adding convolution layer to the mode. 
# Tried 32 filters first and then 64. It is the number of kernels/feature detectors used each
# of shape 3 by 3

# cnn_model.add(Conv2D(32,3, 3, activation='relu'))
# cnn_model.add(MaxPooling2D(pool_size = (2, 2)))

cnn_model.add(Conv2D(64,3, 3, input_shape = (28,28,1), activation='relu'))

# Done with convolution layer and activation function, adding pooling layer to the model
cnn_model.add(MaxPooling2D(pool_size = (2, 2)))


cnn_model.add(Dropout(0.25))

# Adding a flattening layer to flat the features into one single array so that it can be fed into
# fully connected neural network
cnn_model.add(Flatten())

# Hidden layers included between the outputs
cnn_model.add(Dense(output_dim = 32, activation = 'relu'))

# Adding another hidden layer, specifying output dimension as 10, as we need to categorize an 
# image into one of the 10 classes.
# Softmax can also be used for multiclass classification,  but here with 1 class we can use 
#sigmoid in the layer because its range can represent the probability of one class
cnn_model.add(Dense(output_dim = 10, activation = 'sigmoid'))
```

```python
#  Compiling the model, Using Adam optimizer, accuracy as metric
# Using sparse_categorical entropy as we are categorizing data out of 10 classes
# Can use Binary if we had only 0 and 1
cnn_model.compile(loss ='sparse_categorical_crossentropy', optimizer=Adam(lr=0.001),metrics =['accuracy'])
```

```python
# Times the training data will be presented to the model to update the weights
epochs = 50
history = cnn_model.fit(X_train,
                        y_train,
                        batch_size = 512,
                        nb_epoch = epochs,
                        verbose = 1,
                        validation_data = (X_validate, y_validate))
```

# STEP #5: EVALUATING THE MODEL

```python
evaluation = cnn_model.evaluate(X_test, y_test)
# Evaluation returns loss as well
print('Test Accuracy : {:.3f}'.format(evaluation[1]))
```

```python
# get the predictions for the test data
predicted_classes = cnn_model.predict_classes(X_test)
```

```python
L = 5
W = 5
fig, axes = plt.subplots(L, W, figsize = (12,12))
axes = axes.ravel() # 

for i in np.arange(0, L * W):  
    axes[i].imshow(X_test[i].reshape(28,28))
    axes[i].set_title("Prediction Class = {:0.1f}\n True Class = {:0.1f}".format(predicted_classes[i], y_test[i]))
    axes[i].axis('off')

plt.subplots_adjust(wspace=0.5)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, predicted_classes)
plt.figure(figsize = (14,10))
sns.heatmap(cm, annot=True)
# Sum the diagonal element to get the total true correct values
```

```python
from sklearn.metrics import classification_report

num_classes = 10
target_names = ["Class {}".format(i) for i in range(num_classes)]

print(classification_report(y_test, predicted_classes, target_names = target_names))
```
